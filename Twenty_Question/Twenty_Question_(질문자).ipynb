```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

model=ChatOllama(model='llama3.1',temperature=0)

#prompt에 대한 template작성
template=ChatPromptTemplate.from_messages([
    ('system','''당신은 스무고개 게임에서 질문을 하는 사람입니다.
                  Context를 참고하여 Request을 만족하는 문장을 제시하세요.'''),
    ('human','Context: {context}'),
    ('human','Request: {request}')
])
```

```python
prompt=template.invoke({
    'context':''' 다음은 미국에 대한 설명입니다.
1. 미국은 북아메리카 대륙에 위치한 연방 공화국입니다.
2. 북미 대륙 본토에 있는 48개 주와 알래스카, 하와이를 포함한 총 50개의 주, 그리고 특별 행정 구역이자 수도인 워싱턴 D.C.로 구성되어 있습니다.
3. 러시아와 캐나다에 이어 세계에서 세 번째로 넓은 영토를 가지고 있으며, 인구 역시 중국과 인도에 이어 세계 3위 규모를 자랑합니다.
4. 연방 차원에서 지정된 공식 공용어는 없으나 영어가 사실상의 공용어 역할을 하며, 전 세계에서 이주해 온 다양한 인종과 문화가 섞인 '인종의 용광로(Melting Pot)' 사회를 형성하고 있습니다.

다음은 대한민국에 대한 설명입니다.
1. 대한민국은 동아시아의 한반도 남부에 위치한 민주공화국입니다.
2. 북쪽으로는 군사분계선(휴전선)을 경계로 북한과 맞닿아 있으며, 나머지 삼면은 바다로 둘러싸여 있습니다.
3. 수도는 서울이며, 공식 언어로는 독창적인 표기 문자인 한글을 사용하는 한국어를 채택하고 있습니다.
''',
    'request':'미국과 대한민국을 구분하고, "네" 혹은 "아니요"로 답할 수 있는 하나의 질문'
})
```

```python
model.invoke(prompt)
```
